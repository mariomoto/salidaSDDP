```python
import datetime
def my_print(msg: str) -> str:
    now = datetime.datetime.now()
    nowstr = now.strftime("%Y-%m-%d %H:%M:%S,") + f"{now.microsecond // 1000:03d}"
    return f"{nowstr} - {msg}"

my_print("hola")
```
